### Soal 1

Create a query to answer the following questions:
- How many tables are in the austin_bikeshare dataset?

````python
query = """
SELECT
  count(*) AS tables_count
FROM `bigquery-public-data.austin_bikeshare.__TABLES__`
"""
# Set up the query
query_job = client.query(query).to_dataframe()

query_job
````

---
- How many columns in the bikeshare_stations table have STRING data?

````python
query = """
SELECT
  data_type, COUNT(data_type)
FROM `bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.COLUMNS`
WHERE table_name = 'bikeshare_stations'
GROUP BY data_type
HAVING data_type = 'STRING'
"""
# Set up the query
query_job = client.query(query).to_dataframe()

query_job
````

### Soal 2

Create a query to filter data based on the following conditions:
- Retrieve all trips where the bike_id is not null and the duration is greater than 1000 minutes.


````python
query = """
SELECT
  column_name, data_type
FROM `bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.COLUMNS`
WHERE table_name = 'bikeshare_trips'
"""
# Set up the query
query_job = client.query(query).to_dataframe()

query_job
````

````python
query = """
SELECT
  *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE bike_id IS NOT NULL AND duration_minutes > 1000;
"""
# Set up the query
query_job = client.query(query).to_dataframe()

query_job
````

---
- Filter trips where the subscriber_type is "Walk Up" and the end_station_name  is "Stolen"

````python
query = """
SELECT
  *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE subscriber_type = 'Walk Up' AND end_station_name = 'Stolen';
"""
# Set up the query
query_job = client.query(query).to_dataframe()

query_job
````
